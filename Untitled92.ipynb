## Q1. What is the purpose of forward propagation in a neural network?

```python
Input Layer: The input features are fed into the input layer of the neural network. Each input neuron corresponds to a feature in the input data.

Hidden Layers: The input data is propagated forward through one or more hidden layers of the network. In each hidden layer, the input is transformed by applying a set of weights to the input values and adding a bias term. Then, an activation function is applied to the weighted sum of inputs to introduce non-linearity into the network.

Output Layer: Finally, the transformed input is propagated through the output layer of the network. The output layer typically consists of one or more neurons, each representing a possible output class or regression value. The activation function applied to the neurons in the output layer depends on the nature of the task (e.g., sigmoid for binary classification, softmax for multi-class classification, linear for regression).

Output Prediction: The output values generated by the neurons in the output layer represent the predictions made by the neural network for the given input data. These predictions can be compared to the true labels or target values to compute the loss or error, which is used to update the network's parameters during the training process.
```
